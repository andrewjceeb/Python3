```python
asx = pd.read_csv('https://www.asx.com.au/asx/research/ASXListedCompanies.csv',header=1) #CSV reads the URL containing the desired CSV
print(asx)
```

```python
asx_np = np.array(asx) #Converts the asx dataframe into a numpy formatted array
```

```python
asx_name = asx_np[:,0]
asx_tickers = asx_np[:,1]
asx_gics = asx_np[:,2]

sectors = []
for i in range(0,len(asx_tickers)):
        if asx_gics[i] not in sectors:
            sectors.append(asx_gics[i])    
#print(sectors)
#print('')

temp = []
count = 0
for j in range (0,len(sectors)):
    for i in range(0,len(asx_tickers)):
        if asx_gics[i] == sectors[j]:
            print(i, asx_name[i], asx_tickers[i] + ".AX", asx_gics[i])
            count = count + 1
    temp.append(count)
    print('')
#    print('All companies in the ' + sectors[j] + ' GICS sector have been printed')
#    print('')

#print(temp)
counters = [temp[0]]
print('All ' + str(temp[0]) + ' companies in the ' + sectors[0] + ' GICS sector have been printed' )
for k in range(1,len(temp)):
    print('All ' + str(temp[k]-temp[k-1]) + ' companies in the ' + sectors[k] + ' GICS sector have been printed')
    counters.append(temp[k]-temp[k-1])
```

```python
plt.scatter(sectors,counters)
plt.grid(True)
plt.xticks(rotation=90)
plt.ylabel('# of Companies Per GICS Sector')
plt.title('Distribution of Companies By GICS Within the ASX')
plt.show()
```
